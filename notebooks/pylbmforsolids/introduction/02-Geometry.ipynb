Pylbm for solids.

Pylbm is a python-package for numerical simulations using the Lattice-Boltzmann-Method (LBM). For more general information, refer to the website https://pylbm.readthedocs.io/en/latest/, this documentation will focus on the scope of the project, which is to extend and use pylbm to solve equations of quasi-static linear elasticity in 2D. 

Here we will look at how to define more complex geometries.

Pylbm lets us build geometries in a constructive geometry fashion where we can add and substract simple shapes. Available shapes and syntax for all elements can be found in `pylbm/elements/`. Generally we give some geometric parameters for the position/shape/size of the element, define labels for the edges (which we need to specify boundary conditions) and choose if the nodes inside this element should be inside (`isfluid=True`) or outside (`isfluid=False`) the domain. These elements are then gathered in a list and provided to pylbm with the `'elements'` key. The order of this list matters of course, with every element updating the boundary labels and in-or-out flags of the nodes they cover. Below is an example where we successively build a more complex geometry of a ring with a triangle in the center. 

```python
import pylbm

#First we create a box and let all nodes be outside the domain by subtracting a rectangle with the same size as the box

#Parallelograms are defined with a point ((0,0) here) and two vectors ((1,0) and (0,1)) 
shape_1 = pylbm.Parallelogram((0,0), (1, 0), (0, 1), label=-1, isfluid=False)

#create the geometry
dico = {
    'box' : {'x': [0,1], 'y': [0,1], 'label': -1},
    'elements': [shape_1, ]
}
geom = pylbm.Geometry(dico)
geom.visualize(viewlabel = True)
```

```python
import pylbm

#Parallelograms are defined with a point ((0,0) here) and two vectors ((1,0) and (0,1)) 
shape_1 = pylbm.Parallelogram((0,0), (1, 0), (0, 1), label=-1, isfluid=False)

#Circles are defined with a center and a radius
shape_2 = pylbm.Circle((0.5,0.5), 0.5, label=1, isfluid=True)
shape_3 = pylbm.Circle((0.5,0.5), 0.25, label=2, isfluid=False)

#create the geometry
dico = {
    'box' : {'x': [0,1], 'y': [0,1], 'label': -1},
    'elements': [shape_1, shape_2, shape_3]
}
geom = pylbm.Geometry(dico)
geom.visualize(viewlabel = True)
```

```python
import pylbm

#Parallelograms are defined with a point ((0,0) here) and two vectors ((1,0) and (0,1)) 
shape_1 = pylbm.Parallelogram((0,0), (1, 0), (0, 1), label=-1, isfluid=False)

#Circles are defined with a center and a radius
shape_2 = pylbm.Circle((0.5,0.5), 0.5, label=1, isfluid=True)
shape_3 = pylbm.Circle((0.5,0.5), 0.25, label=2, isfluid=False)

#Triangles are defined with a point and two vectors
shape_4 = pylbm.Triangle((0.33,0.33),(0.5,0),(0,0.5), label=[3,4,5], isfluid=True)

#create the geometry
dico = {
    'box' : {'x': [0,1], 'y': [0,1], 'label': -1},
    'elements': [shape_1, shape_2, shape_3, shape_4]
}
geom = pylbm.Geometry(dico)
geom.visualize(viewlabel = True)
```
